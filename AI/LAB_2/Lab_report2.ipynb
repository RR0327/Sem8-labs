8-Puzzle using BFS

````python
# Import deque from collections for efficient queue operations
from collections import deque  # deque = Double-ended queue, fast queue operation er jonno use hoy

# --- Check if state is goal ---
# Function to check if the current state is the same as the goal state
def is_goal(state, goal):
    return state == goal

# --- Find blank position ---
# Function to find the position of the blank tile (represented by 0)
def find_blank(state):
    for i in range(3):  # Loop through rows
        for j in range(3):  # Loop through columns
            if state[i][j] == 0:  # If the cell contains blank (0)
                return i, j  # Return its position (row, column)

# --- Generate successors ---
# Function to generate all possible next states by moving the blank tile
def get_successors(state):
    successors = []  # List to store next states
    x, y = find_blank(state)  # Get blank tile position
    moves = [(-1,0), (1,0), (0,-1), (0,1)]  # Possible moves: Up, Down, Left, Right

    for dx, dy in moves:  # Try each move
        nx, ny = x + dx, y + dy  # New position of blank
        if 0 <= nx < 3 and 0 <= ny < 3:  # Ensure move is within bounds
            # Convert state to a mutable list of lists
            new_state = [list(row) for row in state]
            #print(new_state)
            # Swap blank with the neighboring tile
            new_state[x][y], new_state[nx][ny] = new_state[nx][ny], new_state[x][y]
            # Convert back to tuple of tuples (so it can be stored in a set)
            #print(successors)
            successors.append(tuple(tuple(row) for row in new_state))
    return successors  # Return all possible next states

# --- BFS Search ---
# Function to perform Breadth-First Search to find the solution path
def bfs(start, goal):
    # Initialize queue with the start state and its path
    queue = deque([(start, [start])])
    visited = set()  # Set to keep track of visited states
    visited.add(start)  # Mark start state as visited

    while queue:  # While there are states to explore
        state, path = queue.popleft()  # Get the first state and its path
        if is_goal(state, goal):  # If goal is reached
            return path  # Return the solution path
        for succ in get_successors(state):  # Generate next states
            if succ not in visited:  # If not already visited
                visited.add(succ)  # Mark as visited
                queue.append((succ, path + [succ]))  # Add new state with updated path
    return None  # Return None if no solution is found


# --- Initial and Goal States ---
# Define the starting configuration of the puzzle
initial_state = (
    (1, 2, 3),
    (4, 0, 5),
    (6, 7, 8)
)

# Define the goal configuration of the puzzle
goal_state = (
    (1, 2, 3),
    (4, 5, 6),
    (7, 8, 0)
)

# --- Solve ---
# Run BFS to find a solution from initial to goal state
solution = bfs(initial_state, goal_state)

# --- Print Solution ---
# If a solution is found, display it step by step
if solution:
    print(f"Solution found in {len(solution)-1} moves:")
    for step in solution:  # Print each step in the solution path
        for row in step:  # Print each row of the puzzle
            print(row)
        print()  # Blank line between steps
else:
    print("No solution found.")  # If no solution exists
````

- **What is state?**

A **state** means one arrangement of the puzzle.

Example:
```python
(
    (1, 2, 3),
    (4, 0, 5),
    (6, 7, 8)
)
```
This is one state.

- **What is state space?**

All possible puzzle configurations together are called state space.

In 8-puzzle, there are many possible states.

- **What is successor?**

A successor is a new state generated from current state by one valid move.

Example:

Blank moves right:
```python
(1, 2, 3)
(4, 0, 5)
(6, 7, 8)
```
becomes:
```python
(1, 2, 3)
(4, 5, 0)
(6, 7, 8)
```
- **Why use BFS?**

BFS explores level by level.

So it finds the **shortest path** in terms of number of moves.

- **Why use queue?**

BFS needs queue because it follows:
```bash
First In, First Out
```
The state added first is explored first.

- **Why use visited set?**

To avoid visiting same puzzle state repeatedly.

Without visited:
```bash
The blank may move left, then right, then left again forever.
```
So **visited** prevents infinite repetition.

- **Why use tuple instead of list?**

Because tuple can be stored in a set.

List cannot be stored in set because list is changeable.

So:
```python
visited.add(state)
```
works only when state is tuple.

8-Puzzle using DFS

````python
# =====================================================
# 8-Puzzle Problem using DFS
# =====================================================

# --- Check if state is goal ---
def is_goal(state, goal):
    return state == goal


# --- Find blank position ---
def find_blank(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return i, j


# --- Generate successors ---
def get_successors(state):
    successors = []
    x, y = find_blank(state)

    # Possible moves: Up, Down, Left, Right
    moves = [(-1, 0), (1, 0), (0, -1), (0, 1)]

    for dx, dy in moves:
        nx, ny = x + dx, y + dy

        if 0 <= nx < 3 and 0 <= ny < 3:
            new_state = [list(row) for row in state]

            # Swap blank tile with neighbor tile
            new_state[x][y], new_state[nx][ny] = new_state[nx][ny], new_state[x][y]

            # Convert list back to tuple
            successors.append(tuple(tuple(row) for row in new_state))

    return successors


# --- DFS Search ---
def dfs(start, goal):
    stack = [(start, [start])]
    visited = set()

    while stack:
        state, path = stack.pop()

        if is_goal(state, goal):
            return path

        if state not in visited:
            visited.add(state)

            for successor in get_successors(state):
                if successor not in visited:
                    stack.append((successor, path + [successor]))

    return None


# --- Print Solution ---
def print_solution(solution):
    if solution:
        print(f"Solution found in {len(solution)-1} moves:")

        for step in solution:
            for row in step:
                print(row)
            print()
    else:
        print("No solution found.")


# --- Initial State ---
initial_state = (
    (1, 2, 3),
    (4, 0, 5),
    (6, 7, 8)
)

# --- Goal State ---
goal_state = (
    (1, 2, 3),
    (4, 5, 6),
    (7, 8, 0)
)


# --- Run DFS ---
solution = dfs(initial_state, goal_state)
print_solution(solution)

# --- Final Line ---
if solution:
    print(f"Solution found in {len(solution)-1} moves.")
else:
    print("No solution found.")
````

8-Puzzle using IDS

````python
# =====================================================
# 8-Puzzle Problem using IDS
# =====================================================

# --- Check if state is goal ---
def is_goal(state, goal):
    return state == goal


# --- Find blank position ---
def find_blank(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return i, j


# --- Generate successors ---
def get_successors(state):
    successors = []
    x, y = find_blank(state)

    # Possible moves: Up, Down, Left, Right
    moves = [(-1, 0), (1, 0), (0, -1), (0, 1)]

    for dx, dy in moves:
        nx, ny = x + dx, y + dy

        if 0 <= nx < 3 and 0 <= ny < 3:
            new_state = [list(row) for row in state]

            # Swap blank tile with neighbor tile
            new_state[x][y], new_state[nx][ny] = new_state[nx][ny], new_state[x][y]

            # Convert list back to tuple
            successors.append(tuple(tuple(row) for row in new_state))

    return successors


# --- DLS Function for IDS ---
def dls(start, goal, limit):
    return recursive_dls(start, goal, limit, [start])


def recursive_dls(state, goal, limit, path):
    if is_goal(state, goal):
        return path

    if limit == 0:
        return None

    for successor in get_successors(state):
        # Avoid repeated state in the current path
        if successor not in path:
            result = recursive_dls(successor, goal, limit - 1, path + [successor])

            if result is not None:
                return result

    return None


# --- IDS Search ---
def ids(start, goal, max_depth):
    for depth in range(max_depth + 1):
        print(f"Checking depth limit: {depth}")

        result = dls(start, goal, depth)

        if result is not None:
            return result

    return None


# --- Print Solution ---
def print_solution(solution):
    if solution:
        print(f"Solution found in {len(solution)-1} moves:")

        for step in solution:
            for row in step:
                print(row)
            print()
    else:
        print("No solution found.")


# --- Initial State ---
initial_state = (
    (1, 2, 3),
    (4, 0, 5),
    (6, 7, 8)
)

# --- Goal State ---
goal_state = (
    (1, 2, 3),
    (4, 5, 6),
    (7, 8, 0)
)


# --- Run IDS ---
max_depth = 30
solution = ids(initial_state, goal_state, max_depth)
print_solution(solution)
````
